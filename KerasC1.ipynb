```python
##new train
X_train, X_cv, y_train, y_cv = train_test_split(
    df_features,
    df_lables,
    random_state=1212,
    test_size=0.2
)

X_train = X_train.to_numpy().astype("float32")
X_cv = X_cv.to_numpy().astype("float32")
X_test = X_test.astype("float32")   # if already numpy

X_train /= 255.0
X_cv /= 255.0
X_test /= 255.0
# #Conver lables into one hot encoded
num_digit = 10
y_train = keras.utils.to_categorical(y_train,num_digit)
y_cv = keras.utils.to_categorical(y_cv,num_digit)
```

```python
#printing after encoding
print(y_train[6])
print(y_train[5])
```

```python
#input parameter
n_input = 784
n_hidden1 = 300
n_hidden2= 100
n_hidden3 = 100
n_hidden4 = 200
num_digit = 10
```

```python
inp = Input(shape=(784,))
x = Dense(n_hidden1, activation= "relu", name="hidden_layer_1")(inp)
x = Dense(n_hidden2, activation= "relu", name="hidden_layer_2")(x)
x = Dense(n_hidden3, activation= "relu", name="hidden_layer_3")(x)
x = Dense(n_hidden4, activation= "relu", name="hidden_layer_4")(x)
output = Dense(num_digit, activation= "softmax", name="Output_layer")(x)
```

```python
model = Model(inp,output)
model.summary()
```

```python
#inserting the hyperparameter
learning_rate= 0.001
batch_size = 100
training_epochs = 20
sgd = optimizers.SGD(learning_rate=learning_rate)
```

```python
#Compile the model
model.compile(loss = "categorical_crossentropy",
             optimizer = sgd,
             metrics = ["accuracy"])
```

```python
history1 = model.fit(
    X_train,y_train,
    batch_size= batch_size,
    epochs = training_epochs,
    verbose = 2,
    validation_data = (X_cv,y_cv)
)
```
