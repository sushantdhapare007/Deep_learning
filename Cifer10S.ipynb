```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from keras import models
from keras.models import Model
from keras.datasets import cifar10 
from keras.layers import Conv2D, Input, Dense, Dropout, MaxPooling2D, Flatten
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping
```

```python
from keras import backend as K
K.clear_session()
```

```python
cifer = cifar10.load_data()
```

```python
(X_train,y_train),(X_test,y_test) = cifer
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
#normalization (this normalization is for the pixel value 255 -> 0 or 1)
X_train = X_train.astype("float32")
X_train = X_train / 255
X_test = X_test.astype("float32")
X_test = X_test / 255
```

```python
X_train.shape
```

```python
print(X_train.min(), X_train.max())
print(X_test.min(), X_test.max())
```

```python
#One-Hot Encoding 
y_train = to_categorical(y_train,10)
y_test = to_categorical(y_test,10)
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
model  = models.Sequential()
model.add(Conv2D(filters=(32),kernel_size=(3,3),activation="relu",input_shape=(32,32,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.5))
model.add(Conv2D(filters=(64),kernel_size=(3,3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=(75),kernel_size=(3,3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128,activation = "relu"))
model.add(Dense(10,activation = "softmax"))
```

```python
model.summary()
```

```python
earlystop = EarlyStopping(
    monitor = "val_accuracy",
    patience = 3,
    restore_best_weights = True 
)
```

```python
model.compile(loss = "categorical_crossentropy",
            metrics = ["accuracy"],
            optimizer = "adam")
```

```python
sushant = model.fit(
    X_train,y_train,
    batch_size = 50,
    epochs = 25,
    verbose = 1,
    validation_data =(X_test,y_test),
    callbacks = [earlystop]
)
```
